```python
from bs4 import BeautifulSoup
import selenium
import pymongo
from pymongo.errors import PyMongoError
import requests
from requests import RequestException
import random
import re
from time import sleep
```

```python
client = pymongo.MongoClient()
wino_db = client.wino_db
wines = wino_db.wines
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.113 Safari/537.36'
headers = {'User-Agent':user_agent}
```

```python
page = 1
product_links = []
ret = requests.get('https://specsonline.com/product-category/wine/', headers= headers)
while "Oh Snap!" not in ret.text:
    random_sleep = random.uniform(.002,2)
    soup = BeautifulSoup(ret.text)
    product_links += soup.find_all('a', class_="woocommerce-LoopProduct-link")
    print("Page: " + str(page) + ", Links: " + str(len(product_links)))
    sleep(random_sleep)
    page += 1
    ret = requests.get('https://specsonline.com/product-category/wine/page/' + str(page) + '/', headers=headers)
    
```

```python
formatted_product_links = [link.get('href') for link in product_links]

with open("specs-wine-links.txt", 'w') as oFile:
    oFile.writelines([link + "\n" for link in formatted_product_links])
```

```python
with open("specs-wine-links.txt", 'r') as iFile:
    formatted_product_links = iFile.readlines()
```

```python
for i ,wine_link in enumerate(formatted_product_links):
    
    wine_dict = {}
    wine_link = wine_link[:-1]
    wine_dict['url'] = wine_link
    
    try:
        ret = requests.get(wine_link, headers = headers)
    except RequestException as re:
        print(re)
        continue
    
    soup = BeautifulSoup(ret.text)
    if not soup:
        print("No Soup: " + wine_link)
        continue
    
    title = soup.find('h1', class_='product_title')
    if not title or not title.text:
        print("No Title: " + wine_link)
        continue
    wine_dict['title'] = title.text
    
    specs_img_src = soup.find('div', class_='product').img
    if specs_img_src:
        wine_dict['specs_image_src'] = specs_img_src.get('src')
        
    specs_case_price = soup.find('p', class_='caseprice')
    if specs_case_price:
        wine_dict['specs_case_price'] = specs_case_price.text
    
    specs_bottle_price = soup.find('p', class_="price")
    if specs_bottle_price:
        wine_dict['specs_bottle_price'] = specs_bottle_price.text
     
    specs_summary = soup.find('div', class_='summary').find('p', class_=None)
    if specs_summary and not "Prices shown" in specs_summary.text:
        wine_dict['specs_desc'] = specs_summary.text
    else:
        wine_dict['specs_desc'] = ""
    
    specs_sku = soup.find('span', class_='sku')
    if specs_sku:
        wine_dict['specs_sku'] = specs_sku.text
        
    specs_category = soup.find('span', class_='posted_in').find('a')
    if specs_category:
        wine_dict['specs_category'] = specs_category.text
        
    wine_dict['specs_tags'] = [tag.text for tag in soup.find('span', class_='tagged_as').find_all('a')]
    for row in soup.find('table', class_='shop_attributes').find_all('tr'):
       wine_dict[row.th.text] = row.td.text.strip('\n')
    
    sleep(random.uniform(.002,2))
    try:
        if not wines.find_one({'url':wine_dict['url']}):
            wines.insert_one(wine_dict)
            print("Inserted: " + wine_dict['title'] + ", " + str(i + 1) + " of " + str(len(formatted_product_links)))
        else:
            print("Not Inserted: " + wine_dict['title'] + ", " + str(i + 1) + " of " + str(len(formatted_product_links)))
    except PyMongoError as pme:
        print(pme)
        continue
```
